## **Operadores Matemáticos**

Os operadores matemáticos básicos, tais como: adição, subtração, multiplicação, divisão, exponeciação e radiação podem ser feitos dentro do Python.

```python
# Adição
n1 = 5.887
n2 = 7.15
soma = n1 + n2

# Modos de Apresentação
print('A soma dos dois números é:', soma)
print(f"A soma dos dois números é:, {soma:.2f}")  # 2f é para mostrar as duas casas decimais após a virgula
n1 + n2

```

```python
# Subtração
a1 = 154.587
a2 = 10.548
sub = a1 - a2

# Modos de Apresentação
print("A subtração é", round(sub,2))
print(f"A subtração é {sub:.2f}")
```

**Observação:**

Regras de arredondamento seguida pelo **round()**

Arredondamento para cima (arredondamento por excesso): Se a parte decimal for igual ou maior que 0.5, o número é arredondado para cima. Por exemplo, 2.7 é arredondado para 3.

Arredondamento para baixo (arredondamento por defeito): Se a parte decimal for menor que 0.5, o número é arredondado para baixo. Por exemplo, 2.3 é arredondado para 2.

Arredondamento para o número par mais próximo (arredondamento por paridade): Se a parte decimal for igual a 0.5, o número é arredondado para o número par mais próximo. Por exemplo, 2.5 é arredondado para 2, enquanto 3.5 é arredondado para 4.

Arredondamento para zero (arredondamento em direção a zero): A parte decimal é simplesmente descartada, sem qualquer arredondamento. Por exemplo, 2.9 é arredondado para 2.

Arredondamento para um número específico de casas decimais: O número é arredondado para o número desejado de casas decimais. Por exemplo, arredondar 2.3333 para duas casas decimais resulta em 2.33.

```python
# Multiplicação
b1 = 3.57
multi = b1 * 3

# Modo de apresentação
print(multi)
print(f'A multi é: {multi:.2f}')
```

## **Operadores Lógicos**

Os operadores lógicos em Python são usados para realizar operações lógicas entre expressões booleanas ou valores booleanos.

Eles são comumente utilizados em estruturas de controle condicional, como instruções if e loops, para tomar decisões com base em condições lógicas.

Python possui três operadores lógicos principais:

*  and (E lógico): Retorna True se todas as expressões booleanas forem verdadeiras, caso contrário, retorna False.
*   or (OU lógico): Retorna True se pelo menos uma das expressões booleanas for verdadeira, caso contrário, retorna False.
*   not (NÃO lógico): Inverte o valor de uma expressão booleana. Retorna True se a expressão for falsa e False se a expressão for verdadeira.

Além desses operadores, Python também possui operadores de comparação (==, !=, >=) que retornam valores booleanos. Esses operadores são frequentemente usados em conjunto com operadores lógicos.








Operador And ou '&'

 Retorna verdadeiro se todas as condições forem verdadeiras

```python
teste8 = 3 >= 2 and 3 != 2 and 3 < 4
print(f'3 >= 2 and 3 != 2 and 3 < 4 é: {teste8}')

teste9 = (3 >= 2) & (3 != 2) & (3 < 4)          # use parênteses no &
print(f'3 >= 2 & 3 != 2 & 3 < 4 é: {teste9}')

```

Operador OR ou '|'

Retorna verdadeiro se pelo menos um dos elementos for verdadeiro

```python
teste10 = 3 >= 2 or 3 == 2 or 3 > 4
teste15 = (3 >= 2) | (3 == 2) | (3 > 4)
print(f'3 >= 2 or 3 == 2 or 3 > 43 >= 2 or 3 == 2 or 3 > 4 é: {teste10}')
```

Operador (not) Inverte verdadeiro com Falso

```python
teste11 = not(teste10)  # invertendo o resultado da comparação de todas as sentenças
print(f'not(3 >= 2 or 3 == 2 or 3 > 4) é: {teste11}')
```

```python
teste13 = ((not(3 <= 2))) & (not(3 == 2)) & (3 < 4) # invertendo o resultado da primeira sentença e da segunda sentença
#               V                 V            V
print(f'((not(3 <= 2))) & (not(3 == 2)) & (3 < 4) é: {teste13}')
```

## **Operadores de identidade (is, is not)**

Em Python, os operadores de identidade são utilizados para comparar a identidade de objetos, ou seja, verificar se dois objetos se referem ao mesmo local de memória.

(is) compara se um objeto é igual ao outro em relação ao valor e o tipo de dado

```python
m = 3.0
n = 3
teste18 = m == n
print(f'm == n é: {teste18}')
teste19 = m is n
print(f'm is n é: {teste19}')
print(type(m))
print(type(n))
```

```python
a = "Olá"
b = "Olá"
c = a

print(a is b)  # por mais que tenha o mesmo valor ainda são objetos diferentes
print(a is c)

print(type(a))
print(type(b))
```

(is not) compara se um objeto é diferente do outro em relação ao valor e o tipo

```python
o = 3
s = 3.0
teste22 = o != s
print(f'{o} é diferente em relação a valor do {s}: {teste22}')

teste23 = o is not s
print(f'{o} é diferente em relação a valor e tipo de dados do {s}: {teste23}')

print(type(o))
print(type(s))
```

Instruçõa "else":



*   A instrução "else" é usada em conjunto de instruções "if" e especifica em um bloco de código a ser executado se a condição do "if" não for verdadeira. A sintaxe geral é a seguinte:



```python
'''
if condição:
  # bloco de código a ser executado se a condição for verdadeira
else:
  # o bloco de código a ser executado se a condição não for verdadeira
'''
```

```python
idade = int(input('Digite sua idade: '))

if idade >= 18:
  print('Maior de idade')
else:
  print('Menor de idade')
```

Instrução "elif":

* A instrução "elif" é uma abreviação de "else if" e permite testar múltiplas condições em sequência. Ela é usada quando há mais de duas possibilidades a serem consideradas. A sintexa geral é:



```python
'''
if condição:
  # bloco de código a ser executado se a condição for verdadeira
elif condição:
  # bloco de código a ser executado se a condição 2 for verdadeira
else:
  # o bloco de código a ser executado se a condição não for verdadeira
'''
```

```python
nota = float(input('Digite sua nota: '))

if nota >= 90:
  print('sua nota é A')
elif nota >= 80:
  print('sua nota é B')
elif nota >= 70:
  print('sua nota é C')
else:
  print('Sua nota é D')
```

Exemplo:
*  Escreva um programa que verifique se um número digitado pelo usuário é positivo, negativo ou zero.

Passo a passo:

1. Solicite ao usuário que digite um número.
2. Armazene o número digitado em uma variável chamada "numero".
3. Verifique se o número digitado é igual a zero. Se sim, exiba a mensagem "O número digitado é zero."
4. Caso contrário, verifique se o número é maior que zero. Se sim, exiba a mensagem "O número digitado é positivo."

 Se verdadeiro imprima a mensagem aprovado e sua nota

 Se falso imprima reprovado e sua nota

```python
 # Recebe uma lista com várias notas
notas = [5, 9, 6, 8, 7]

for i, v in enumerate(notas): # primeira variável é o índice e a segunda variável é o valor
# se uma nota qualquer encontrada for maio ou igual a 7 faça algo
  if notas[i] >= 7:
    print(f'aprovado com nota {notas[i]}')
  else:
    print(f'reprovado com nota {notas[i]}')
  print(notas[i])
```

```python
n1 = input('Informe uma sequência de n números inteiros separados por espaços: ')
n2 = input('''Informe outra sequência de números inteiros separados por espaços
com a mesma quantidade informada na lista anterior''')

# Criação de duas listas: X e Y
x = n1.split()
y = n2.split()

# For para converter os valores das listas para float
for i in range(len(x)):
  x[i] = float(x[i])
  y[i] = float(y[i])

# Criação das listas vazias
xquad = []
yquad = []
xy = []

# For para elevar os valores das listas ao quadrado
for i in range(len(x)):
  xquad.append(x[i]**2)
  yquad.append(y[i]**2)
  xy.append(x[i]*y[i])

# Criando as variáveis que armazenam a soma
sx = sum(x)
sy = sum(y)
sxquad = sum(xquad)
syquad = sum(yquad)
sxy = sum(xy)

# Apresentação das somas
print(f'A soma dos valores de x é {sx}')
print(f'A soma dos valores de y é {sy}')
print(f'A soma dos valores de xquad é {sxquad}')
print(f'A soma dos valores de yquad é {syquad}')
print(f'A soma dos valores de xy é {sxy}')
```

Elabore uma função que receba em seis inputs diferentes as seguintes informações:
1. Receita líquida em R$

2. Custo em R$

3. Estoques em R$

4. Contas a receber em R$

5. Fornecedores em R$

6. O Número de dias do balanço (int)

Após receber as informações acima a função deve calcular o valor e printá-lo dos
seguintes KPIs:

a) Prazo médio de estoques - PME

b) Prazo médio de Recebimento - PMR

c) Prazo médio de Pagamento PMP

d) Ciclo operacional - CO

e) Ciclo de caixa CC

E por último:

Se PMP for maior que ciclo operacional print a mensagem essa em média empresa
paga tantos dias depois de receber

Se PMP for menor que ciclo operacional print a mensagem essa em média empresa
paga tantos dias antes de receber

Se PMP for igual ao ciclo operacional print a mensagem essa em média empresa
no mesmo dia em que recebe


```python
def giro ():
  receita = float(input('Informe a receita líquida em R$: '))
  custo = float(input('Informe o custo em R$: '))
  estoques = float(input('Informe o estoque em R$: '))
  ctas = float(input('Informe as contas a receber em R$: '))
  forne = float(input('Informe os fornecedores em R$: '))
  n = int(input('Informe o período do balanço em dias inteiros: '))
  pme = (estoques/custo)*n
  pmr = (ctas/receita)*n
  pmp = (forne/custo)*n
  co = pme + pmr
  cc = co - pmp
  print(f'O Prazo médio de estoques é  aproximadamente {pme:.0f} dias'),
  print(f'O Prazo médio de recebimento é  aproximadamente {pmr:.0f} dias'),
  print(f'O Prazo médio de pagamento é  aproximadamente {pmp:.0f} dias'),
  print(f'O ciclo operacional é  aproximadamente {co:.0f} dias'),
  print(f'O Ciclo de caixa é  aproximadamente {cc:.0f} dias')
  if pmp > co:
    print(f'Essa empresa em média paga {abs(cc):.0f} dias depois de receber')
  elif pmp < co:
    print(f'Essa empresa em média paga {abs(cc):.0f} dias antes de receber')
  else:
    print('Essa em média empresa no mesmo dia em que recebe')
```
